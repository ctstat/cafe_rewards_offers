```python
import ast
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 

offers = pd.read_csv('C:/Users/ctsta/Desktop/cafe_rewards_offers/offers.csv')
offers.head()
```

```python
offers['num_channels'] = offers.channels.apply(lambda x: len(ast.literal_eval(x)))
offers.head(10)
```

# Convert String to List in Python Using ast.literal_eval()

```python
# disclosure feature data type (string not a list)
a = offers['channels'][0]
a
```

```python
# Examples
# import ast module:  Abstract Syntax Trees
import ast

# Example1: outter with single quote and inner with double duote
a_string = '["apple", "orange", "watermelon"]'
a_list = ast.literal_eval(a_string)
print(a_list)

# Example2: outter with double quote and inner with single duote
b_string = "['black', 'red', 'white']"
b_list = ast.literal_eval(b_string)
print(b_list)
```

```python
offer2 = offers.copy()
offer2['channels'] = offer2.channels.apply(lambda x: ast.literal_eval(x))
offer2.head()
```

# one-hot encoded channels 

```python
# toy example
df = pd.DataFrame({'id':range(1,4), 
                   'items':[['A', 'B'], ['A', 'B', 'C'], ['A', 'C']]})
df
```

```python
df.join(df['items'].str.join(sep='*').str.get_dummies(sep='*'))
```

```python
# save the results
offer_clean = offer2.join(offer2['channels'].str.join(sep='*').str.get_dummies(sep='*'))
offer_clean.to_csv('C:/Users/ctsta/Desktop/cafe_rewards_offers/clean_offers.csv', index = False)
```
